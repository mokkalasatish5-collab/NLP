```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp("Elon flew to mars yesterday. He carried biryani masala with him")
```

```python
for token in doc:
    print(token," | ", token.pos_, " | ", spacy.explain(token.pos_))
```

```python
doc = nlp("Wow! Dr. Strange made 265 million $ on the very first day")

for token in doc:
    print(token," | ", token.pos_, " | ", spacy.explain(token.pos_))
```

```python
doc = nlp("Wow! Dr. Strange made 265 million $ on the very first day")

for token in doc:
    print(token," | ", token.pos_, " | ", spacy.explain(token.pos_), " | ", token.tag_, " | ", spacy.explain(token.tag_))
```

```python
doc = nlp("He quits the job")

print(doc[1].text, "|", doc[1].tag_, "|", spacy.explain(doc[1].tag_))
```

```python
with open('new_story.txt','r') as f:
    file_data = f.read()
file_data
```

```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
text = nlp(file_data)
```

```python
Noun_list = []
for token in text:
    if token.pos_ == 'NOUN':
        Noun_list.append(token.text)
print("Noun list:",Noun_list)
```

```python
num_pos = []
for token in text:
    if token.pos_ == 'NUM':
        num_pos.append(token.text)
print("Num_pos list:",num_pos)
print(len(num_pos))
```

```python
pos_tags = []
for token in text:
    pos_tags.append(token.tag_)
print(set(pos_tags))
print(len(set(pos_tags)))
```
