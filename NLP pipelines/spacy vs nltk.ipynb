```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")

doc = nlp("this is satish. how are you?")
for sentence in doc.sents:
    print(sentence)
```

```python
for sentence in doc.sents:
    for word in sentence:
        print(word)
```
