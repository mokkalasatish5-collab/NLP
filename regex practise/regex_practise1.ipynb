```python
def get_pattern_match(pattern,text):
    pat = re.findall(pattern,text)
    return pat

get_pattern_match(r'Born\s*.*\n(.*)\(age',text)

```

```python
def get_details(text):
    name = get_pattern_match(r'Born\s*(.*)',text)
    age = get_pattern_match(r'age (\d)',text)
    birthyear = get_pattern_match(r'Born\s*.*\n(.*)\(age',text)
    place = get_pattern_match(r'\(age.*\n(.*)\n',text)

    return{
        'name':name[0],
        'age':int(age[0]),
        'birthyear':birthyear[0],
        'place':place[0]
    }
```
